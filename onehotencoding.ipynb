```python
#sklearn has two classes, transformers and estimators(numeric before ingesting using one hot encording)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe_df = pd.DataFrame(ohe.fit_transform(df).toarray())
ohe_df.head()
```

```python
pd.DataFrame(ohe.fit_transform(df[['smoker']]).toarray()).head()
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder()
ct = make_column_transformer(
    (ohe, ['smoker', 'sex', 'day', 'time']),
    remainder='passthrough'
)
```

```python
data_new = pd.DataFrame(ct.fit_transform(df))
```
